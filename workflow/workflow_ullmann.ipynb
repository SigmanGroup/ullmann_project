# HELPER FUNCTIONS

```python
def process_dataframe(input_df):
    """
    Process a DataFrame by generating new column names and removing the first two rows.

    Parameters:
    - input_df: DataFrame to be processed.

    Returns:
    - processed_df: Processed DataFrame with updated column names and removed first two rows.
    """
    # Create a copy of the input DataFrame
    processed_df = input_df.copy()

    # Generate new column names based on the first two rows
    column_names = [
        processed_df.iloc[1][col] if col in processed_df.columns[0:2]
        else processed_df.iloc[0][col] + "_" + processed_df.iloc[1][col]
        for col in processed_df.columns
    ]

    # Assign the new column names to the DataFrame
    processed_df.columns = column_names

    # Skip the first two rows
    processed_df = processed_df.iloc[2:]

    # Reset the index
    processed_df.reset_index(drop=True, inplace=True)

    # Keep only specific columns based on their names
    columns_to_keep = [
        col for col in processed_df.columns
        if col.endswith("Name") or col.endswith("smiles") or col.endswith("min") or col.endswith("max")
    ]
    processed_df = processed_df[columns_to_keep]

    return processed_df

def normalize_dataframe(dataframe, variance_threshold=0.0):
    """
    Normalize specified columns in a DataFrame using Min-Max scaling and apply variance thresholding.

    Parameters:
    - dataframe: DataFrame to be normalized.
    - variance_threshold: Threshold for variance-based feature selection.

    Returns:
    - normalized_df: Normalized DataFrame with selected columns.
    """
    # Create a copy of the input DataFrame
    normalized_df = dataframe.copy()

    # Extract columns to normalize
    columns_to_normalize = dataframe.columns[2:]

    # Apply Min-Max scaling to the specified columns
    scaler = MinMaxScaler()
    normalized_df[columns_to_normalize] = scaler.fit_transform(dataframe[columns_to_normalize])

    # Apply variance thresholding to select relevant columns
    selector = VarianceThreshold(threshold=variance_threshold)
    selected_columns = selector.fit_transform(normalized_df[columns_to_normalize])
    
    # Combine the selected columns with the original non-normalized columns
    normalized_df = normalized_df.iloc[:, :2].join(pd.DataFrame(selected_columns, columns=columns_to_normalize[selector.get_support()]))

    return normalized_df

def correlated_pairs(dataframe, threshold=0.95):
    """
    Identify highly correlated column pairs in a DataFrame based on a correlation threshold.

    Parameters:
    - dataframe: DataFrame to analyze.
    - threshold: Correlation coefficient threshold for identifying pairs.

    Returns:
    - None
    """
    # Create a correlation matrix and extract upper triangular part
    df_corr = dataframe.corr().abs()
    upper = df_corr.where(np.triu(np.ones(df_corr.shape), k=1).astype(bool))

    # Find pairs of highly correlated columns
    correlated_pairs = []
    for col1 in upper.columns:
        for col2 in upper.columns:
            if col1 != col2 and upper[col1][col2] > threshold:
                correlated_pairs.append((col1, col2))

    # Print the correlated pairs if found
    if correlated_pairs:
        print(f"Highly correlated column pairs with a correlation coefficient above {threshold}:")
        for col1, col2 in correlated_pairs:
            print(f"Correlated Pair: {col1} : {col2}")
    else:
        print("No highly correlated column pairs found.")

    return None

def drop_columns(dataframe, substrings):
    """
    Drop columns in a DataFrame containing specified substrings.

    Parameters:
    - dataframe: DataFrame to be processed.
    - substrings: List of substrings to identify columns for dropping.

    Returns:
    - dataframe: DataFrame with specified columns dropped.
    """
    # Identify columns containing specified substrings and drop them
    columns_to_drop = [col for col in dataframe.columns if any(substring in col for substring in substrings)]
    return dataframe.drop(columns=columns_to_drop)
```

# DECISION TREE

## TRAINING THE DECISION TREE

```python
# Create a copy of the training DataFrame 'data'
data = training_df.copy()

# Define the starting column index for properties and the response column index
properties_start = 5
response = 4

# Extract features (X) and target (y) from the DataFrame
X = data.iloc[:, properties_start:].values
y = data["Yield"]

# Convert yield values to binary classes (0: 0-20%; 1: 20%+)
y_class = np.array([0 if i < 20 else 1 for i in y])
y_class = y_class.astype(int)

# Get the names of the features (X)
X_labelname = data.columns[properties_start:].tolist()

# Count the number of reactions in each class
count_zero = np.count_nonzero(y_class == 0)
print(f"There are {count_zero} reactions with yields between 0-20%")
count_ones = np.count_nonzero(y_class == 1)
print(f"There are {count_ones} reactions with yields 20% or above")

# Plot a histogram to visualize the distribution of classes
fig, ax = plt.subplots(figsize=(3, 2))
plt.hist(y_class, bins=[-0.1, 0.1, 0.9, 1.1])
plt.xlabel('Yield \n[0: 0-20%; 1: 20%+]')
plt.ylabel('Frequency')
plt.xticks([0, 1])
plt.show()
```

### GRID SEARCH TO FIND OPTIMAL HYPERPARAMETERS

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_class, test_size=0.25, shuffle=True, random_state=random_state)

# Set up KFold cross-validation with 4 splits
kfold = KFold(n_splits=4, shuffle=True, random_state=random_state)

# Define the parameter grid for the Decision Tree classifier
param_grid = {
    'criterion': ['gini', 'entropy'],
    'class_weight': [{0: 1, 1: 3}, {0: 1, 1: 4}, {0: 1, 1: 5}, {0: 1, 1: 6}],
    'max_depth': [2, 3, 4, 5]
}

# Perform grid search using DecisionTreeClassifier and specified parameter grid
grid_search = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=kfold, return_train_score=True)
grid_search.fit(X_train, y_train)

# Print the best parameters and cross-validation score
print("Best parameters: {}".format(grid_search.best_params_))
print("Best cross-validation score: {:.2f}".format(grid_search.best_score_))

# Create a DataFrame to store the results of the grid search
results_grid = pd.DataFrame(grid_search.cv_results_)
```

### FINAL DECISION TREE

```python
# Define the class weight for the Decision Tree model
class_weight = {0: 1.0, 1: 3.0}

# Create a Decision Tree model with specified hyperparameters and class weight
model_final = DecisionTreeClassifier(
    random_state=random_state,
    max_depth=3,
    criterion='entropy',
    min_samples_split=3,
    min_samples_leaf=3,
    max_features=None,
    class_weight=class_weight
)

# Fit the Decision Tree model to the training data
model_final.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model_final.predict(X_test)

# Calculate and print evaluation metrics (Accuracy, Precision, Recall, F1 Score)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print("Accuracy: {:.2f}".format(accuracy))
print("Precision: {:.2f}".format(precision))
print("Recall: {:.2f}".format(recall))
print("F1 Score: {:.2f}".format(f1))

# # Plot the confusion matrix
# fig, ax = plt.subplots(figsize=(2, 1.8))
# cm = confusion_matrix(y_test, y_pred, normalize='true')
# cm = np.flipud(cm)

# yticklabels = list(reversed(['Off', 'On']))
# res = sns.heatmap(cm, annot=True, cmap='Greys', fmt='.1%', cbar=False, yticklabels=yticklabels)
# for _, spine in res.spines.items():
#     spine.set_visible(True)
#     spine.set_linewidth(0.75)

# ax.set_xlabel('Predicted Yield')
# ax.set_ylabel('True Yield', font="Arial")
# ax.set_xticks([0.5, 1.5], ['Off', 'On'])
# ax.set_yticks([0.5, 1.5], ['On', 'Off'])

# fig.tight_layout()
# # fig.savefig("cm.png", dpi=600, bbox_inches='tight')
# plt.show()

# Visualize the Decision Tree using graphviz
dot_data = export_graphviz(model_final, out_file=None,
                           feature_names=X_labelname,
                           class_names=['0-20%', '20%+'],
                           filled=True, rounded=True,
                           special_characters=True)

graph = graphviz.Source(dot_data)
graph
```

```python
# Set up KFold cross-validation with 4 splits
kfold = KFold(n_splits=4, shuffle=True, random_state=random_state)

# Calculate and print the average cross-validation score
scores = cross_val_score(model_final, X, y_class, cv=kfold)
average_score = scores.mean()
print("Average cross-validation score: {:.2f}".format(average_score))

# Perform k-fold cross-validation and store confusion matrices
confusion_matrices = []
for train_idx, test_idx in kfold.split(X):
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y_class[train_idx], y_class[test_idx]

    # Train the model on the training data
    model_final.fit(X_train, y_train)

    # Make predictions on the test data
    y_pred = model_final.predict(X_test)

    # Calculate the confusion matrix for this fold
    cm = confusion_matrix(y_test, y_pred, normalize='true')

    # Append the confusion matrix to the list
    confusion_matrices.append(cm)

# Create a figure with subplots for each confusion matrix
fig, axes = plt.subplots(1, len(confusion_matrices), figsize=(2 * len(confusion_matrices), 1.8))
yticklabels = ('Off', 'On')

# Enumerate over the confusion matrices and plot them in the subplots
for i, cm in enumerate(confusion_matrices):
    ax = axes[i]
    res = sns.heatmap(cm, annot=True, cmap='Greys', fmt='.1%', cbar=False, yticklabels=yticklabels, ax=ax)
    for _, spine in res.spines.items():
        spine.set_visible(True)
        spine.set_linewidth(0.75)

    ax.set_xlabel('Predicted Yield')
    ax.set_ylabel('True Yield', font="Arial")
    ax.set_xticks([0.5, 1.5], ['Off', 'On'])
    ax.set_yticks([0.5, 1.5], ['Off', 'On'])

fig.tight_layout()
plt.show()
```

#### Leave-one-compound-out accuracy test

```python
compounds = ["P100", "P101", "P102", "P103", "P104", "P105", "P106", "P107", "P108", "P109",
             "P110", "P111", "P112", "P113", "P114", "P115", "P116", "P117", "P118", "P119",
             "P120", "P121", "P122", "P123", "P124", "P125", "P126", "P127"]

# Create an empty list to store the accuracy scores
loo_scores = []
compound_scores = []

for compound in compounds:
    # Exclude the current compound from the dataset
    leave_one_out_df = training_df[training_df['Product'] != compound]
    
    # Split the dataset into features (X) and target (y)
    X_loo = leave_one_out_df.iloc[:, 5:].values
    y_loo = leave_one_out_df["Yield"]
    y_class_loo = np.array([0 if i < 20 else 1 for i in y_loo])

    # Use the already trained model to predict on the left-out compound
    y_pred_loo = model_final.predict(X_loo)

    # Evaluate the model and store the accuracy score
    accuracy = accuracy_score(y_class_loo, y_pred_loo)
    loo_scores.append(accuracy)

    # Include only the current compound from the dataset
    compound_df = training_df[training_df['Product'] == compound]
    # Split the dataset into features (X) and target (y)
    X_onecmpd = compound_df.iloc[:, 5:].values
    y_onecmpd = compound_df["Yield"]
    y_class_onecmpd = np.array([0 if i < 20 else 1 for i in y_onecmpd])

    # Use the already trained model to predict on the left-out compound
    y_pred_onecmpd = model_final.predict(X_onecmpd)

    # Evaluate the model and store the accuracy score
    accuracy_onecmpd = accuracy_score(y_class_onecmpd, y_pred_onecmpd)
    compound_scores.append(accuracy_onecmpd)
    
# Calculate the average accuracy across all folds
average_loo = np.mean(loo_scores)
average_onecmpd = np.mean(compound_scores)

print("Average Accuracy LOO: {:.2f}".format(average_loo))
print("Average Accuracy One Compound:{:.2f}".format(average_onecmpd))

pd.set_option('display.max_rows', None)
accuracy_df = pd.DataFrame({
    'Product': compounds,
    'Leave-One-Compound-Out Accuracy': loo_scores,
    'Product-Only Accuracy': compound_scores
})
display(accuracy_df)
```

# CONFIDENCE MAPS

## ENTROPY AND CONFIDENCE

### PREPARING DATAFRAME

```python
confidence = result_df.copy()

# Define prefixes to find in column names
prefixes_to_find = ['Br_82_', 'Am_55_', 'Lig_2_']

# Select columns based on prefixes
selected_columns = confidence.columns[0:5].tolist()
found_columns = [col for col in confidence.columns if any(col.startswith(prefix) for prefix in prefixes_to_find)]
selected_columns.extend(found_columns)

# Update confidence DataFrame with selected columns
confidence = result_df[selected_columns]

# Create a binary column based on yield threshold
confidence.insert(5, 'yield-binary', np.where(confidence['Yield'] < 20, 0, 1))

# Set threshold values
threshold_1 = 2.068
threshold_2 = -0.803
threshold_3 = 33.509

# Define column names for convenience
Am = am_name
Br = br_name
Lig = lig_name

# Filter the DataFrame based on ligand distance threshold
filtered_dt = confidence[confidence[Lig] <= threshold_1].copy()

# Add a 'Prediction' column based on additional thresholds
filtered_dt["Prediction"] = 0
filtered_dt.loc[(filtered_dt[Am] <= threshold_2) & (filtered_dt[Br] <= threshold_3), "Prediction"] = 1

# Display the filtered DataFrame
# display(filtered_dt)
```

### COUNTING CORRECT CLASSIFICATIONS

```python
# Grouping by 'Product' and aggregating the counts
counts = filtered_dt.groupby('Product')['yield-binary'].value_counts().unstack()

# Renaming the columns for clarity
counts.columns = ['count_0', 'count_1']

# Resetting the index to make 'Product' a regular column
counts = counts.reset_index()

# Merging the counts DataFrame with the original DataFrame
merged_df = pd.merge(filtered_dt.drop('yield-binary', axis=1), counts, on='Product', how='left')

# Dropping duplicates and keeping the first occurrence of each product
merged_df = merged_df.drop_duplicates(subset='Product', keep='first')

# Reordering the columns
merged_df = merged_df[['Product', Br, Am,'Prediction', 'count_0', 'count_1']]

# Replacing NaN values with 0.001
merged_df = merged_df.fillna(0.001)

# Printing the updated DataFrame
display(merged_df)
```

### COMPUTING ENTROPY

```python
# Calculate total count
merged_df['Total_Count'] = merged_df['count_0'] + merged_df['count_1']        

# Iterate through rows to calculate entropy
for index, row in merged_df.iterrows():
    total_count = row['Total_Count']
    count_0 = row['count_0']
    count_1 = row['count_1']

    if total_count > 0:
        p_0 = count_0 / total_count
        p_1 = count_1 / total_count

        # Calculate entropy based on prediction
        if row['Prediction'] == 0:
            if count_0/total_count >= 0.5:
                entropy = (p_0 * math.log2(p_0))
                merged_df.loc[index, 'Entropy'] = (-1 - entropy) * 100
            else:
                merged_df.loc[index, 'Entropy'] = (-p_0) * 100

        elif row['Prediction'] == 1:   
            if count_1/total_count >= 0.5:
                entropy = -(p_1 * math.log2(p_1))
                merged_df.loc[index, 'Entropy'] = (1 - entropy)*100
            else:
                merged_df.loc[index, 'Entropy'] = (p_1) * 100

        else:
            entropy = 0.0  # Invalid prediction, set entropy to 0.0

    else:
        merged_df.loc[index, 'Entropy'] = 0.0

# Reset the index and display the DataFrame
merged_df.reset_index(drop=True, inplace=True)
display(merged_df)

# Save the DataFrame to an Excel file (optional)
# merged_df.to_excel('confidence.xlsx', index=False)
```

```python
# Set the maximum number of displayed rows to 10 for better readability
pd.set_option('display.max_rows', 10)

# Select specific columns for k-nearest neighbors (knn) analysis
columns_knn = ["Product", "Ligands", "Yield", br_name, am_name]
knn_df = training_df[columns_knn].copy()

# Normalize bromine (Br) and amine (Am) columns using Min-Max scaling
scaler = MinMaxScaler()
knn_df['Br_Norm'] = scaler.fit_transform(knn_df[[br_name]])
knn_df['Am_Norm'] = scaler.fit_transform(knn_df[[am_name]])

# Drop original amine and bromine columns as they are now normalized
knn_df.drop(columns=[am_name, br_name], inplace=True)

# Sort the DataFrame by "Product" and "Yield" in descending order
knn_df = knn_df.sort_values(by=["Product", "Yield"], ascending=[True, False])

# Create a dictionary to store the top 5 yields for each product
top_5_dict = {}
for product, group in knn_df.groupby("Product"):
    top_5_yields = group["Yield"].head(5).tolist()
    top_5_dict[product] = top_5_yields

# Mark rows with a top 5 yield for the respective product
knn_df["top_5"] = knn_df.apply(lambda row: 1 if row["Yield"] in top_5_dict[row["Product"]] else 0, axis=1)

# Reset the index and keep only rows with a top 5 yield
knn_df = knn_df[knn_df["top_5"] == 1]
knn_df = knn_df.reset_index(drop=True)

# Drop the temporary "top_5" column as it's no longer needed
knn_df.drop(columns="top_5", inplace=True)

# Display the resulting DataFrame with top 5 yields for each product
display(knn_df)
```

```python
# Create a copy of the DataFrame for further processing
clean_df = knn_df.copy()

# Remove the "Yield" column from the copied DataFrame
clean_df.drop(columns=["Yield"], inplace=True)

# Create a new column for ligand ranks within each product group
clean_df['Lig_Rank'] = clean_df.groupby('Product').cumcount() + 1

# Pivot the DataFrame to have Lig_1, Lig_2, Lig_3 columns
pivoted_df = clean_df.pivot(index='Product', columns='Lig_Rank', values='Ligands').reset_index()

# Combine the pivoted DataFrame with the "Br_Norm" and "Am_Norm" columns
pivoted_df[['Br_Norm', 'Am_Norm']] = clean_df.groupby('Product')[['Br_Norm', 'Am_Norm']].first().reset_index()[['Br_Norm', 'Am_Norm']]
pivoted_df.rename(columns={
    'Br_Norm': 'Br_Norm',
    'Am_Norm': 'Am_Norm',
    1: 'Top_1',
    2: 'Top_2',
    3: 'Top_3',
    4: 'Top_4',
    5: 'Top_5'
}, inplace=True)
pivoted_df = pivoted_df.dropna(axis=1)

# Create a separate DataFrame "prod_info_df" for product information
prod_info_df = pivoted_df.copy()

# Initialize Nearest Neighbors with 3 neighbors
neighbors = NearestNeighbors(n_neighbors=3)
neighbors.fit(prod_info_df[['Br_Norm', 'Am_Norm']])

# Find the distances and indices of the nearest neighbors
distances, nearest_neighbor_idx = neighbors.kneighbors(prod_info_df[['Br_Norm', 'Am_Norm']])

# Create new columns for the first and second nearest neighbors
prod_info_df['NN_1'] = prod_info_df.loc[nearest_neighbor_idx[:, 1], 'Product'].values
prod_info_df['NN_2'] = prod_info_df.loc[nearest_neighbor_idx[:, 2], 'Product'].values

# Display the resulting DataFrame with ligand information and nearest neighbors
display(prod_info_df)
```

```python
# Create a copy of the DataFrame for further processing
new_df = prod_info_df.copy()

# Define a function to check if ligands are present in the nearest neighbors
def check_ligands_in_nn(row, nn_column):
    nn_product = row[nn_column]
    nn_row = prod_info_df[prod_info_df["Product"] == nn_product]
    if nn_row.empty:
        return False
    
    nn_ligands = set(nn_row[["Top_1", "Top_2", "Top_3"]].values[0])
    row_ligands = set(row[["Top_1", "Top_2", "Top_3", "Top_4", "Top_5"]].values)
    return any(ligand in nn_ligands for ligand in row_ligands)

# Apply the function to create new columns for checking ligands in nearest neighbors
new_df["Check_NN_1"] = new_df.apply(lambda row: check_ligands_in_nn(row, "NN_1"), axis=1)
new_df["Check_NN_2"] = new_df.apply(lambda row: check_ligands_in_nn(row, "NN_2"), axis=1)

# Combine the results using logical OR to create a new column "check"
new_df["check"] = new_df["Check_NN_1"] | new_df["Check_NN_2"]

# Count the number of True values in the "check" column
true_count = new_df["check"].sum()

# Display the DataFrame with the new columns
print("Number of True values in the 'check' column:", true_count)
pd.set_option('display.max_rows', None)
display(new_df)

# Save the DataFrame to an Excel file
# new_df.to_excel('knn.xlsx', index=False)
```
