# IMPORTS

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib
import matplotlib.patches as patches

from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from matplotlib.ticker import MultipleLocator

matplotlib.rcParams['font.family'] = 'Arial'
matplotlib.rcParams['font.size'] = 10
```

# LOADING DATASET

```python
####### Load expt and DFT data
excel = "../computed_data"
expt_sheet = "expt_data"
lig_sheet = "DFT_lig"

####### Make the loaded data into dataframes
expt_df = pd.read_excel(excel + ".xlsx", expt_sheet, header=None, engine='openpyxl')
ligands_df = pd.read_excel(excel+".xlsx",lig_sheet,header=None, engine='openpyxl')
# display(expt_df)
# display(ligands_df)
```

```python
####### Adding headers
headers = pd.concat([ligands_df.iloc[0:2, 2:]], axis=1)
row_idx = [0, 1]
col_idx = range(5, 5 + len(headers.columns))
expt_df.loc[row_idx, col_idx] = headers.values
# display(expt_df)

####### Searching for start idx
ligands_start = None
# Loop over the column names in expt
for idx in expt_df.iloc[0].items():
    if "Lig_1" in idx:
        ligands_start = idx[0]

####### Appending features to each ligand
for idx, row in expt_df.iterrows():
    if idx > 1:
        cell_lig = row[3]
        
        # Find the row in br dataframe that matches the cell_value
        lig_row = ligands_df[ligands_df.iloc[:, 0] == cell_lig]

        # Check if there is only one row that matches the cell_value
        if len(lig_row) == 1:
            # Get the columns 1 to end of the row in br dataframe
            lig_cols = lig_row.iloc[0, 2:]

        # Iterate over each column in br_cols and assign the values to corresponding columns in expt
            for col_idx, col_value in enumerate(lig_cols[0:], start=ligands_start):
                expt_df.at[idx, col_idx] = col_value

pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', 20)
# display(expt_df.iloc[:, :5])

display(expt_df)
```

# CLEANING FEATURES

```python
def select_columns(column_name):
    if column_name in expt_df.columns[0:5]:
        return True
    elif expt_df.iloc[1][column_name].endswith("_min") or expt_df.iloc[1][column_name].endswith("_max"):
        return True
    else:
        return False

selected_columns = [col for col in expt_df.columns if select_columns(col)]
clean_df = expt_df[selected_columns].copy()

column_names = [expt_df.iloc[1][col] if col in expt_df.columns[0:5] else expt_df.iloc[0][col] + "_" + expt_df.iloc[1][col] for col in clean_df.columns]
clean_df.columns = column_names

clean_df = clean_df.iloc[2:]
clean_df.reset_index(drop=True, inplace=True)

display(clean_df)
```

# INITIAL PROD/LIGANDS AND FINAL PROD/LIGANDS

```python
initial_df = clean_df[(clean_df['Product'].str[1:].astype(int) <= 111) & (clean_df['Ligands'].str.split('_').str[1].astype(int) <= 24)]
new_df = clean_df[(clean_df['Product'].str[1:].astype(int) <= 111) & (clean_df['Ligands'].str.split('_').str[1].astype(int) > 24)]
all_df = clean_df[(clean_df['Product'].str[1:].astype(int) < 200)]

pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', 15)

display(initial_df)
display(new_df)
# display(all_df)
```

# DF PREPARATION

```python
initial_df.loc[:, 'Yield'] = pd.to_numeric(initial_df['Yield'], errors='coerce')
grouped = initial_df.groupby("Ligands")
max_yield_indices = grouped["Yield"].idxmax()

initial_lig_df = initial_df.loc[max_yield_indices]

initial_lig_df.reset_index(drop=True, inplace=True)

display(initial_lig_df)

#############################
new_df.loc[:, 'Yield'] = pd.to_numeric(new_df['Yield'], errors='coerce')
grouped_new = new_df.groupby("Ligands")
max_yield_new = grouped_new["Yield"].idxmax()

new_lig_df = new_df.loc[max_yield_new]

new_lig_df.reset_index(drop=True, inplace=True)

display(new_lig_df)
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', 15)
# Get the first row
first_row = initial_lig_df.iloc[[0]]

# Drop the first row from the original DataFrame
initial_lig_df = initial_lig_df.drop(initial_lig_df.index[0])

# Append the first row to the end of the DataFrame
initial_lig_df = initial_lig_df.append(first_row)
display(initial_lig_df)
display(new_lig_df)
```

# THRESHOLD INVESTIGATION

```python
def plot_decision_tree_feature(X, y_class, X_labelname, class_weight, min_accuracy=0.90):
    plot_colors = ['#E63946', '#1D3557']
    colors = ['#E63946', '#1D3557']
    cmap = ListedColormap(colors)

    for i, f_ind in enumerate(X_labelname):
        dt = DecisionTreeClassifier(max_depth=1, class_weight=class_weight).fit(X[:, i].reshape(-1, 1), y_class)
        y_pred = dt.predict(X[:, i].reshape(-1, 1))
        accuracy = accuracy_score(y_class, y_pred)
        f1 = f1_score(y_class, y_pred, average='macro')
        
        if accuracy > min_accuracy:
            print(f"Feature: {f_ind}")
            print(f"Accuracy: {accuracy:.2f}")
            print(f"F1 score: {f1:.2f}")
            print(f"Decision threshold: {dt.tree_.threshold[0]:.2f}")
            
            dt_plt = DecisionTreeClassifier(max_depth=1, class_weight=class_weight).fit(X[:, i].reshape(-1, 1), y_class)
            x_min, x_max = X[:,i].min(), X[:,i].max()
            y_min, y_max = 0, 100
            
            dx, dy = x_max - x_min, y_max - y_min
            plot_step = dx / 100
            xx, yy = np.meshgrid(np.arange(x_min - 0.04 * dx, x_max + 0.04 * dx, plot_step),
                                 np.arange(y_min - 0.04 * dy, y_max + 0.04 * dy, plot_step))
            
            plt.figure(figsize=(3, 3))
            Z = dt.predict(xx.ravel().reshape(-1, 1))
            Z = Z.reshape(xx.shape)
            
            plt.contourf(xx, yy, Z, cmap=cmap, alpha=0.1)
            
            for j, color in zip(range(2), plot_colors):
                idx = np.where(y_class == j)
                plt.scatter(X[idx, i], y[idx], c=color, edgecolor="Black", linewidth=0.75, s=25)
            plt.axhline(y=20, color='grey', linestyle='dotted')
            
            plt.xlabel("X")
            plt.ylabel("Y")
            plt.tight_layout()
            
            plt.show()
            print("\n----------------------------------------------\n")
```

```python
data = initial_lig_df

X_labelname = data.columns[5:].tolist()
X = data.iloc[:, 5:].values
y = data["Yield"].values
y = y.astype(float)
y_class = np.array([0 if i < 20 else 1 for i in y])
n_classes = 2
class_weight = {0:1,1:20}

min_accuracy_threshold = 0.95  # Adjust the value as needed
plot_decision_tree_feature(X, y_class, X_labelname, class_weight, min_accuracy=min_accuracy_threshold)
```
