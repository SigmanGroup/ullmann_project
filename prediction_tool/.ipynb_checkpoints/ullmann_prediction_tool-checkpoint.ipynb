# LOADING DATASET

```python
# Specify the Excel file name (without extension)
excel = "../computed_data"

# Define sheet names for different datasets
br_sheet = "arbr_prediction"
amine_sheet = "amine_prediction"
lig_sheet = "DFT_lig"

# Read the aryl-bromide prediction data into a DataFrame (br_df)
br_df = pd.read_excel(excel + ".xlsx", br_sheet, header=0)

# Read the primary amine prediction data into a DataFrame (amine_df)
amine_df = pd.read_excel(excel + ".xlsx", amine_sheet, header=0)

# Read the ligands data from the DFT calculations into a DataFrame (ligands_df)
# Use only specified columns "Compound_Name" and "smiles"
ligands_df = pd.read_excel(excel + ".xlsx", lig_sheet, header=1, usecols=["Compound_Name", "smiles"])
```

# SELECTION OF PRIMARY AMINE AND ARYL-BROMIDE

```python
# Primary amine input loop
while True:
    # Prompt the user to enter a SMILES string for a primary amine
    amine_input = input("Enter SMILES string of a primary amine in the library: ")
    
    # Convert the entered SMILES to isomeric SMILES
    amine_input = convert_to_isomeric_smiles(amine_input)
    
    # Initialize amine_pred to an empty list
    amine_pred = []
    
    try:
        # Try to find the Nùõø‚Äì value for the entered isomeric SMILES in the amine DataFrame
        amine_pred = amine_df.loc[amine_df['smiles'] == amine_input, 'Nùõø‚Äì'].values[0]
        print(f"The Nùõø‚Äì value for the entered compound: {amine_pred}")
        break  # Exit the loop if a valid SMILES is entered
    except IndexError:
        # Handle the case where the entered SMILES is not found in the amine library
        print("Primary amine not found in the library.")
        retry = input("Do you want to try again? [yes/no]: ")
        if retry.lower() != 'yes' and retry.lower() != 'y':
            print("Exiting the program.")
            break
        else:
            continue  # Ask for input again if the user wants to retry

# Check if amine_pred is a valid numeric value
if not isinstance(amine_pred, (int, float)):
    print("No primary amine entered. Exiting the program.")
    sys.exit()

# Retrieve InChIKey and structure for the entered primary amine
amine_inchi = amine_df.loc[amine_df['smiles'] == amine_input, 'InChIKey'].values[0]
mol_amine = Chem.MolFromSmiles(amine_input)
plt_amine = Chem.Draw.MolToImage(mol_amine, size=(150, 150), sanitize=False)

# Display the structure of the entered primary amine
print("Primary amine structure:")
display(plt_amine)

# Aryl-bromide input loop
while True:
    # Prompt the user to enter a SMILES string for an aryl-bromide
    br_input = input("Enter SMILES string of an aryl-bromide in the library: ")
    
    # Convert the entered SMILES to isomeric SMILES
    br_input = convert_to_isomeric_smiles(br_input)
    
    # Initialize br_pred to an empty list
    br_pred = []
    
    try:
        # Try to find the Steric value for the entered isomeric SMILES in the aryl-bromide DataFrame
        br_pred = br_df.loc[br_df['smiles'] == br_input, 'Steric'].values[0]
        print(f"The Steric value for the entered compound: {br_pred}")
        break  # Exit the loop if a valid SMILES is entered
    except IndexError:
        # Handle the case where the entered SMILES is not found in the aryl-bromide library
        print("Aryl-bromide not found in the library.")
        retry = input("Do you want to try again? [yes/no]: ")
        if retry.lower() != 'yes' and retry.lower() != 'y':
            print("Exiting the program.")
            break
        else:
            continue  # Ask for input again if the user wants to retry

# Check if br_pred is a valid numeric value
if not isinstance(br_pred, (int, float)):
    print("No aryl-bromide entered. Exiting the program.")
    sys.exit()

# Retrieve InChIKey and structure for the entered aryl-bromide
br_inchi = br_df.loc[br_df['smiles'] == br_input, 'InChIKey'].values[0]
mol_br = Chem.MolFromSmiles(br_input)
plt_br = Chem.Draw.MolToImage(mol_br, size=(150, 150), sanitize=False)

# Display the structure of the entered aryl-bromide
print("Aryl-bromide structure:")
display(plt_br)

print(f'\n----------------------------------------------------------------------------------------------------\n')
```

# PREDICTION TOOL

```python
# Convert predictions to NumPy arrays
X = np.array([amine_pred])
Y = np.array([br_pred])

# Normalize the target coordinates
x_target_norm = (X - loaded_x_values[0]) / (loaded_x_values[1] - loaded_x_values[0])
y_target_norm = (Y - loaded_y_values[0]) / (loaded_y_values[1] - loaded_y_values[0])

# Stack the normalized coordinates to create input data
input_data = np.column_stack((x_target_norm, y_target_norm))

print(f'\n----------------------------------------------------------------------------------------------------\n')
# Make predictions with the confidence model
confidence_pred = confidence_model(x_target_norm, y_target_norm)

# Display prediction results
if confidence_pred[0] < 0:
    print(f'This combination of aryl-bromide and primary amine is predicted to yield <20% with {-confidence_pred[0]:.1f}% confidence.')
else: 
    print(f'This combination of aryl-bromide and primary amine is predicted to yield >20% with {confidence_pred[0]:.1f}% confidence.')

print(f'\n----------------------------------------------------------------------------------------------------\n')

# Find the nearest neighbors and their distances
distances, indices = knn_model.kneighbors(input_data)

# Define product IDs for the first and second nearest neighbors
P1 = f'P{100 + indices[0][0]}'
P1_info = training_dict[P1]
P2 = f'P{100 + indices[0][1]}'
P2_info = training_dict[P2]

# Create a meshgrid for contour plotting
x_grid, y_grid = np.meshgrid(np.linspace(0, 1, 500), np.linspace(0, 1, 500))

# Calculate confidence values on the meshgrid
confidence_interp = confidence_model(x_grid, y_grid)

# Clip confidence values to a reasonable range
confidence_interp_norm = np.clip(confidence_interp, -100, 100)

# Extract normalized coordinates of the nearest neighbors
X_NN = np.array([P1_info['Normalized_Nùõø‚Äì'], P2_info['Normalized_Nùõø‚Äì']])
Y_NN = np.array([P1_info['Normalized_Steric'], P2_info['Normalized_Steric']])
neighbors_data = np.column_stack((X_NN, Y_NN))

# Plot the confidence contour and data points
plt.figure(figsize=(3, 3))
contour = plt.contourf(x_grid, y_grid, confidence_interp_norm, cmap='RdBu', levels=100)
scatter = plt.scatter(input_data[:, 0], input_data[:, 1], c=confidence_pred, cmap='RdBu', marker='X', s=75, edgecolors='black', linewidth=1, vmin=-1, vmax=1)
scatter = plt.scatter(neighbors_data[:, 0], neighbors_data[:, 1], marker='o', s=10, color='black', linewidth=1)
plt.xlabel('Amine [Normalized N$^\delta$$^‚Äì$]')
plt.ylabel('Aryl-Bromide [Normalized Steric]')
plt.show()

print(f'\n----------------------------------------------------------------------------------------------------\n')

# Display information for the first nearest neighbor
print(f"\nThe first nearest neighbor is {P1}. The top ligands for this product are: ")

# Define top 3 ligands as L1, L2, and L3
L1 = ligands_df.loc[ligands_df['Compound_Name'] == P1_info['Top_1_Ligand'], 'smiles'].values[0]
L2 = ligands_df.loc[ligands_df['Compound_Name'] == P1_info['Top_2_Ligand'], 'smiles'].values[0]
L3 = ligands_df.loc[ligands_df['Compound_Name'] == P1_info['Top_3_Ligand'], 'smiles'].values[0]

# Create a list of ligands, names, and yields for visualization
smiles_list = [ligands_df.loc[ligands_df['Compound_Name'] == P1_info[f'Top_{i}_Ligand'], 'smiles'].values[0] for i in range(1, 4)]
name_list = [P1_info[f'Top_{i}_Ligand'] for i in range(1, 4)]
yield_list = [P1_info[f'Top_{i}_Yield'] for i in range(1, 4)]

# Create a list of Mol objects from SMILES
best_ligands = [Chem.MolFromSmiles(smi) for smi in smiles_list]

# Create images of ligands and display them
plt_ligands = [Chem.Draw.MolToImage(mol, size=(300, 150), sanitize=False) for mol in best_ligands]
fig, axs = plt.subplots(1, 3, figsize=(8, 4))
for i, ax in enumerate(axs):
    ax.imshow(plt_ligands[i])
    ax.text(0.5, -0.1, f"{name_list[i]}\nYield: {yield_list[i]}%", ha='center', va='center', transform=ax.transAxes, fontsize=10)
    ax.axis("off")

plt.show()

print(f"Here are pubchem links for the three ligands:"
        f"\n{P1_info['Top_1_Ligand']}: {ligands[P1_info['Top_1_Ligand']]}"
        f"\n{P1_info['Top_2_Ligand']}: {ligands[P1_info['Top_2_Ligand']]}"
        f"\n{P1_info['Top_3_Ligand']}: {ligands[P1_info['Top_3_Ligand']]}")

print(f'\n----------------------------------------------------------------------------------------------------')

# Display information for the second nearest neighbor
print(f"\nThe second nearest neighbor is {P2}. The top ligands for this product are: ")

# Define top 3 ligands as L1, L2, and L3:
L4 = ligands_df.loc[ligands_df['Compound_Name'] == P2_info['Top_1_Ligand'], 'smiles'].values[0]
L5 = ligands_df.loc[ligands_df['Compound_Name'] == P2_info['Top_2_Ligand'], 'smiles'].values[0]
L6 = ligands_df.loc[ligands_df['Compound_Name'] == P2_info['Top_3_Ligand'], 'smiles'].values[0]

# Create a list of ligands, names, and yields for visualization
smiles_list = [L4, L5, L6]
name_list = [P2_info['Top_1_Ligand'], P2_info['Top_2_Ligand'], P2_info['Top_3_Ligand']]
yield_list = [P2_info['Top_1_Yield'], P2_info['Top_2_Yield'], P2_info['Top_3_Yield']]

# Create a list of Mol objects from SMILES
best_ligands = [Chem.MolFromSmiles(smi) for smi in smiles_list]
plt_ligands = [Chem.Draw.MolToImage(mol, size=(300, 150), sanitize=False) for mol in best_ligands]
fig, axs = plt.subplots(1, 3, figsize=(8, 4))

# Create images of ligands and display them
for i, ax in enumerate(axs):
    ax.imshow(plt_ligands[i])
    ax.text(0.5, -0.1, f"{name_list[i]}\nYield: {yield_list[i]}%", ha='center', va='center', transform=ax.transAxes, fontsize=10)
    ax.axis("off")

plt.show()

print(f"Here are pubchem links for the three ligands:"
        f"\n{P2_info['Top_1_Ligand']}: {ligands[P2_info['Top_1_Ligand']]}"
        f"\n{P2_info['Top_2_Ligand']}: {ligands[P2_info['Top_2_Ligand']]}"
        f"\n{P2_info['Top_3_Ligand']}: {ligands[P2_info['Top_3_Ligand']]}")
```
